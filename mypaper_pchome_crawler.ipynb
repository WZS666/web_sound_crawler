```python
import requests
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from bs4 import BeautifulSoup
```

```python
# 'https://mypaper.pchome.com.tw/ti1080con'
# 這邊是在用假身分(headers)去爬蟲 (但這個網站沒有防爬，比較沒關係)
headers = requests.utils.default_headers()
headers.update(
{
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.2 (KHTML, like Gecko) Chrome/22.0.12.16.0 safari/537.2',
}
)
```

```python
# 這裡是在取網址裡面個子標題網址裡的內容
def get_content(url,headers):
    text = ''
    test = requests.get(url,headers=headers)
    test.encoding='utf-8'
    test_soup = BeautifulSoup(test.text,'lxml')
    for j in test_soup.find_all('div',class_='innertext brk_h',id="imageSection1"):
        for k in j.find_all('p'):
            text += k.text
    return text
```

```python
# 取得文章的標題、網址及內容
def get_artical(artical):
    df = pd.DataFrame(columns = ['title','title_href','title_content'])
    for i in artical:
        title = i.find('a').text
        title_href = "https://mypaper.pchome.com.tw" + i.find('a').get('href')
        title_content = get_content(title_href,headers)
        df = df.append({'title':title,'title_href':title_href,'title_content':title_content},ignore_index=True)
    return df
```

```python
# 設定要爬幾頁網址
def page_get(page):
    all_page_dataframe = pd.DataFrame(columns = ['title','title_href','title_content'])
    for i in range(page):
        # Get website's information
        response = requests.get('https://mypaper.pchome.com.tw/ti1080con/P'+str(i), headers=headers)
        response.encoding='utf-8'
        soup = BeautifulSoup(response.text,'lxml')
        artical = soup.find_all('h3',class_='title brk_h')
        df = get_artical(artical)
        all_page_dataframe = all_page_dataframe.append(df)
    return all_page_dataframe
```

```python
my_df = page_get(2)
```
